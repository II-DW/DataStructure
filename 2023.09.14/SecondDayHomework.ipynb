# 6. 문자형 변수에 누군가의 학점이 주어졌을 경우, 평이 몇 점인지 출력하는 프로그램을 작성하여라. A+은 4.3, A는 4, A-는 3.7, B+는 3.3, B는 3, B-는 2.7, C+는 2.3, C는 2.0, C-는 1.7, D+는 1.3, D는 1.0, D-는 0.7, F는 0.0의 점수를 가진다.

```python
score:str = input()

socreInt:int = 0
if 'A' in score :
  socreInt += 4
elif 'B' in score :
  socreInt += 3
elif 'C' in score :
  socreInt += 2
elif 'D' in score :
  socreInt += 1

if '+' in score :
  socreInt += 0.3
elif '-' in score :
  socreInt -= 0.3

print(socreInt)
```

# 7. 입력받은 양의 숫자까지의 짝수와 홀수를 나누어서 따로 더하고 출력하는 프로그램을 작성하여라.

Cheat Version.

```python
x:int = int(input())
if x % 2 == 0 :
  print(int((x) * (x/2) / 2), int((2 + x) * (x/2) / 2))
else :
  print(int((x + 1) * ((x+1)/2) / 2), int((x + 1) * ((x-1)/2) / 2))
```

Standard Version.

```python
x:int = int(input())
if x % 2 == 0 :
  print(sum([i for i in range (1, x+1, 2)]), sum([i for i in range (2, x+2, 2)]))
else :
  print(sum([i for i in range (1, x+2, 2)]), sum([i for i in range (2, x+1, 2)]))
```

# 9. 누군가의 생년월일을 입력받아 2023년 9월 21일 기준 몇 끼를 먹었는지 계산하는 프로그램을 작성하여라. (단, 태어났을 때부터 하루 3끼씩 먹었다고 가정한다.)

날짜는 그레고리력을 기준으로 함.

```python
year, month, date = map(int,input().split())

monthList =  [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
answer = 0

def isLeapYear (y:int) -> str :
  if y % 4 == 0 : # 윤년
    if y % 100 == 0: # 윤년 X
      if y % 400 == 0 :
        return True
      else :
        return False
    else :
      return True
  else :
    return False

for y in range(year+1, 2023) :
  if isLeapYear(y) :
    answer += 366
  else :
    answer += 365

if isLeapYear(year) and month < 3:
  answer += sum(monthList[month+1:12]) + 1 # 2006년 4월부터 12월
  if month == 1 :
    answer += monthList[date-1] - date + 1
  if month == 2 :
    answer += monthList[date-1] - date + 2
else :
  answer += sum(monthList[month+1:12]) # 2006년 4월부터 12월
  answer += monthList[date-1] - date + 1
answer += sum(monthList[:9]) + 21

print(answer*3)
```
