# 문제1
> 함수 내에서 반복문을 활용하여 입력받은 값의 구구단을 출력하는 프로그램을 구현하자

```python
def multiplication (num:int) -> str :
  prtStr:str = ''
  for i in range (1, 10) :
    prtStr += str(num) + ' * ' + str(i) + ' = ' + str(num*i) +'\n'
  return prtStr

print(multiplication(int(input())))
```

# 예시 2
> 입력값 (input)과 출력값 (output)이 있는 함수 만들기

```python
def number(num) :
  if num < 10 :
    return '한자리수'
  elif num >= 10 and num < 100 :
    return '두자리수'
  else :
    return '세자리수'

x = int(input())
y = number(x)
print(y)
```

# 문제4

> 입력값이 3.14 초과면  
(입력값) is bigger than pi  
입력값이 3.14 미만이면  
(입력값) is smaller than pi  
입력값이 3.14이면  
(입력값) is pi를 출력하는 프로그램을 작성하여라

```python
x:float = float(input())

if x > 3.14 :
  print(x, 'is bigger than pi')
elif x < 3.14 :
  print(x, 'is smaller than pi')
else :
  print(x, 'is pi')
```

# 문제4
> 정수를 하나 입력받아, 1부터 입력받은 수까지 각각에 대해 제곱을 구해 프린트하는 프로그램을 while문을 활용하여 제작해보자.

```python
x:int = int(input())

i = 0
while i < x :
  i+= 1
  print(i**2)
```

# 문제5
> while을 사용하여 입력받은 만큼 자신의 이름을 출력하는 코드를 작성해보자.

```python
x:int = int(input())

i = 0
while x > i :
  i+=1
  print('이도원')
```

# 문제6
