# 시험용

```python
def number(n) :
  if n % 2 == 0:
    print("짝수")
  else :
    print("홀수")

num1 = int(input())
number(num1)
```

```python
n = int(input())
for i in range (1, n+1) :
    print(' ' * (i-1) + '*' * (2*(n-i)+1))
```
