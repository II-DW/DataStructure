# 복습

```python
n = int(input())
for i in range (1, n+1) :
    print(' ' * (n-i) + '*' * (2*i - 1))
```

# 트리

> 트리는 비선형 구조이며, 계층적 자료를 표현하는데 적합하다.

> ex : 폴더, 의사 결정 트리, DFS (너비 우선 탐색), BFS (깊이 우선 탐색)

> 계층적으로 가장 높은 곳에 있는 곳을 루트 트리, 나머지를 서브 트리

> 루트와 루트를 잇는 것, 간선, 자식이 없는 노드를 단말 노드, 자식이 있는 노드를 내부 노드라고 함.

> 깊이 : 층의 개수

> 차수 : 가장 아래 노드의 개수



### 이진 트리

> 서브 트리가 공집합인 루트만 존재하는 트리 역시 이진 트리라고 볼 수 있다.

> 서브 트리 간의 순서가 구별 - 좌우 구별

> 이진 트리의 모든 노드는 차수가 2 이하 즉 자식 노드의 개수가 2 이하이다. 반면 일반 트리는 자식 노드의 개수에 제한이 없다.

> 이진 트리에서 노드는 루트를 제외하면 정확하게 부모 ㅋ 노드의 개수가 n개이면 간선의 개수는 n-1개이다.

> 높이가 h인 이진 트리의 경우, 최소 h개의 노드를 가지며 최대 2^h - 1개의 노드를 가진다. 한 레벨어 적어도 하나의 노드는 존재해야하기 떄문임.

> 하나의 노드는 최대 2개의 자식을 가질 수 있으므로, 레벨 i 에서의 최대 노드 개수는 2^(i-1)이다. 이걸로 전체 노드 개수 구하면 2^h - 1 이 됨.

> 포화 이진 트리 - 각 레벨에 노드가 꽉 차있는 이진 트리,

> 완전 이진 트리는 다른 노드는 모두 꽉 차있지만, 마지막 레벨에서는 왼쪽부터 오른쪽으로 노드가 순서대로 채워져 있는 트리

> 포화 이진 트리는 항상 완전 이진 트리라고 볼 수 있지만, 그 역은 성립하지 않음.

> 트리는 데이터를 저장하기 위한 자료 구조, 순회하며 정보를 찾음
전위순회, 중위순회, 후위순회가 있음.

> 전위 순회
1. 루트 노드를 방문한다.
2. 왼쪽 노드를 방문한다.
3. 만약 왼쪽 노드가 없다면 오른쪽 노드를 방문한다.

<img src='https://upload.wikimedia.org/wikipedia/commons/a/ac/Preorder-traversal.gif'></img>

> 중위 순회
1. 왼쪽 서브 트리를 방문한다.
2. 오른쪽 서브트리를 방문한다.
3. 루트 노드를 방문한다.

<img src='https://upload.wikimedia.org/wikipedia/commons/4/48/Inorder-traversal.gif'></img>

> 후위 순회
1. 왼쪽 서브 트리의 모든 노드를 방문한다.
2. 오른쪽 서브 트리의 모든 노드를 방문한다.
3. 루트 노드를 방문한다.

<img src='https://upload.wikimedia.org/wikipedia/commons/2/28/Postorder-traversal.gif'></img>

> 레벨 순회 :
각 노드를 레벨 순으로 검사하는 순회 방법이다. ( 너비 우선 탐색과 같음)
