# 연결 리스트

> ARRAY라는 문자열의 자리는 6자리

> A 메모리 다음에 다음 정보를 담고있는 메모리의 주소를 저장함

> A 메모리 -> R 메모리 -> ...

> 속도가 빠름

> 커졌다 작아졌다 가능

> 맨 위에 있는 메모리를 Head Node라고 함.

> Data node -> Next node (주솟값)

> 새로운 값을 추가하기도 용이함. (중간에)

> 삭제하기도 용이함. (중간에)

> 원형 연결 리스트는 헤더 파일이 없음.

> 단순 연결 리스트 : 연결 리스트의 노드는 구조체로 생성되며, 동적 메모리를 사용하여 생성하고 해제한다.
