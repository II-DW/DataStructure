# 1. Stack

### 1. LIFO ( Last In First Out ) 형식
 > 먼저 들어올 수록 나중에 나감

### 2. stack은 크기가 정해져있으며, 크기를 초과 했을 때 overflow 현상이 일어남.

### 3. Push
 > TOP = -1 : TOP은 스택을 제어하는 변수, 0부터 시작하기 위해 초깃값은 -1  

 > 데이터가 한개 들어온다면 0, 두 개 들어와있다면 1과 같은식으로 변경

 > stack의 크기가 3으로 정해져있다면, TOP=2일 때, 추가 하려면 overflow 현상이 일어남.



### 4. POP
> 가장 최근에 들어온 변수가 삭제 되게 됨.

### 5. 예시
> 인터넷에서 뒤로가기

> 메모리에서도 사용됨.

# 2. QUEUE

### 1. FIFO ( First In FIrst Out ) 구조
> 먼저 들어온 사람이 먼저 나감

### 2. Enqueue, Dequeue

> Enqueue 데이터가 들어와서 쌓임.

> Dequeue 젤 처음 들어왔던 데이터가 사라짐.

> 먼저 들어온 데이터가 앞으로 차근차근 움직이는 것처럼 보일 수 있음.

### 3. Front
> DATA1이 들어오면, roar( 현재 위치를 알려주는 변수 )가 DATA1의 위치를 가르킴. (0)

> DATA2가 또 들어오면, roar은 한 칸 앞으로 움직여 DATA2 자리를 가르킴. (1)
