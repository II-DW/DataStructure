```python
L = [3, 8, 22, 50, 4, 6]

for i in range (len(L)) :
  for j in range (i, len(L)) :
    if L[j] < L[i] :
      tmp = L[i]
      L[i] = L[j]
      L[j] = tmp
      print(L)
```
